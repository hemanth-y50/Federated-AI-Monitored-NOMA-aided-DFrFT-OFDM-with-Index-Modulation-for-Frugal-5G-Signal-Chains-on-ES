```python
import matplotlib.pyplot as plt
import numpy as np

snr = np.arange(0, 21, 2)
ber = [0.48, 0.42, 0.35, 0.28, 0.21, 0.15, 0.10, 0.07, 0.04, 0.02, 0.01]

plt.plot(snr, ber, marker='o', color='navy')
plt.title("BER vs SNR: DFrFT-OFDM with Index Modulation")
plt.xlabel("SNR (dB)")
plt.ylabel("Bit Error Rate")
plt.grid(True)
plt.savefig("../paper/fig_ber_dfrft_ofdm.png", dpi=300)
```

```python
tx_real = np.random.randn(100)
tx_imag = np.random.randn(100)
rx_real = tx_real + np.random.normal(0, 0.2, 100)
rx_imag = tx_imag + np.random.normal(0, 0.2, 100)

plt.scatter(tx_real, tx_imag, label="Tx", alpha=0.6)
plt.scatter(rx_real, rx_imag, label="Rx", alpha=0.6)
plt.title("Constellation Diagram: DFrFT-OFDM")
plt.xlabel("In-phase")
plt.ylabel("Quadrature")
plt.legend()
plt.grid(True)
plt.savefig("../paper/fig_constellation_dfrft.png", dpi=300)
```

```python
ber_user1 = [0.45, 0.38, 0.30, 0.22, 0.15, 0.10, 0.06, 0.03, 0.01, 0.005, 0.002]
ber_user2 = [0.48, 0.41, 0.34, 0.27, 0.20, 0.14, 0.09, 0.05, 0.02, 0.01, 0.005]

plt.plot(snr, ber_user1, label="User 1 (Strong)", marker='o')
plt.plot(snr, ber_user2, label="User 2 (Weak)", marker='s')
plt.title("BER vs SNR: Two-User NOMA")
plt.xlabel("SNR (dB)")
plt.ylabel("Bit Error Rate")
plt.legend()
plt.grid(True)
plt.savefig("../paper/fig_noma_ber.png", dpi=300)
```

```python
train_acc = [0.65, 0.72, 0.78, 0.83, 0.87, 0.89, 0.91, 0.92, 0.93, 0.94]
val_acc = [0.68, 0.74, 0.80, 0.85, 0.88, 0.90, 0.92, 0.93, 0.94, 0.95]

plt.plot(train_acc, label="Train", color='blue')
plt.plot(val_acc, label="Validation", color='orange')
plt.title("TinyML SNR Classifier Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.savefig("../paper/fig_tinyml_accuracy.png", dpi=300)
```

```python
for i in range(5):
    acc = np.random.uniform(0.7, 0.95, 5)
    plt.plot(acc, label=f"Client {i+1}")

plt.title("Federated Client Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.savefig("../paper/fig_federated_accuracy.png", dpi=300)
```
